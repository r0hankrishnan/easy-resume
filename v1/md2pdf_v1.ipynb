# MD2PDF Version 1 Workflow

In this version, we use a more realistic .md document and implement 3 different themes.

```python
import markdown
from weasyprint import HTML
```

```python
theme_choice = "theme_modern.css"
```

```python
with open(theme_choice, "r", encoding = "utf-8") as f:
    css = f.read()
```

```python
with open("sample_realistic.md", "r", encoding = "utf-8") as f:
    md_text = f.read()
```

```python
html_body = markdown.markdown(md_text, extensions = ["fenced_code", "tables", "toc", "attr_list"])
```

```python
html_full = f"""
<!doctype html>
<html>
<head>
<meta charset="utf-8">
<title>Booklet</title>
<style>{css}</style>
</head>
<body>
{html_body}
</body>
</html>
"""
```

```python
with open("output.html", "w", encoding = "utf-8") as f:
    f.write(html_full)
print(f"Exported with theme: {theme_choice}")
```

```python
HTML("output.html").write_pdf("output.pdf", full_fonts = True)
```
